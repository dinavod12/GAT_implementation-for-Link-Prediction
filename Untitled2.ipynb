```python
# Creating Adjacency Matrix for Edge Feature

from sklearn.preprocessing import StandardScaler
import numpy as np

def adj_func(datas):

    sc = StandardScaler()

    lst1 = [int(a) for a in datas.edge_index[0]]
    lst2 = [int(b) for b in datas.edge_index[1]]

    tensor1 = np.array([lst1])
    tensor2 = np.array([lst2])

    max_index = max(np.max(tensor1), np.max(tensor2))+1

    adj_matrix = np.zeros((3084,3084), dtype=float)



    data_val = sc.fit_transform(datas.edge_attr)

    for i in range(len(datas.edge_index[0])):
        adj_matrix[datas.edge_index[0][i],datas.edge_index[1][i]] = data_val[i][0]


    return adj_matrix,max_index
```

```python
import torch.nn as nn
from torch_geometric.nn import GATv2Conv,GATConv
import torch.nn.functional as F

class New_GAT_two_layer(torch.nn.Module):

  def __init__(self, dim_in, dim_h, dim_out, heads=1):
    super().__init__()
    self.layer1 = GATv2Conv(dim_in, dim_h, heads=heads)
    self.layer2 = GATv2Conv(dim_h*heads, dim_out, heads=1)
    self.update = nn.Linear(dim_h*heads+dim_out,dim_out)

  def encode(self,adj_matrix,x,edge_val):
     gat1 = F.relu(self.layer1(x,edge_val))
     gat2 = self.layer2(gat1,edge_val)
     gat_final = torch.cat((gat1,gat2),1)
     val = F.relu(self.update(gat_final))
     return val
  def decode(self,updated,edge_val):
    val = (updated[edge_val[0]]*updated[edge_val[1]]).sum(dim=-1)

    return val
```

```python
model = New_GAT_two_layer(trainset.num_features,68,34)
optimizer = torch.optim.Adam(model.parameters(),lr=0.01, weight_decay=5e-4)
criterion = torch.nn.BCEWithLogitsLoss()
#criterion = CustomLoss()
#criterion = nn.MSELoss()
```

```python
# Calling Adjacency matrix function for train data

train_adj_matrix,train_max_val = adj_func(trainset)
```

```python
# Function for Normalization using MinMax Normalization

def func_normalize(data_value):
  from sklearn.preprocessing  import MinMaxScaler
  mm = MinMaxScaler()
  train_x = mm.fit_transform(data_value)
  return train_x
```

```python
# Creating negative edge for train data

train_neg_edge_index = neg_edge_index[:, train_indices]
neg_edge_labels = torch.zeros(neg_edge_index.size(1))
```

```python
train_edge_label_index = torch.cat(
            [trainset.edge_index, train_neg_edge_index],
            dim=-1,
        )

train_edge_label = torch.cat([
            trainset.edge_labels,
            neg_edge_labels[train_indices]
        ], dim=0)
```

```python
train_loss = []
import torch.nn.functional as F
def train():
  model.train()
  optimizer.zero_grad()
  z = model.encode(torch.Tensor(train_adj_matrix),torch.Tensor(func_normalize(trainset.x)),trainset.edge_index)
  out = model.decode(z, train_edge_label_index).view(-1)
  #loss=criterion(out,train_edge_label)
  loss = F.binary_cross_entropy_with_logits(out,train_edge_label)
  loss.backward()
  optimizer.step()
  train_loss.append(loss.item())
  return loss.item()
```

```python
# Calling Adjacency matrix function for test data

test_adj_matrix,test_max_val = adj_func(testset)
```

```python
# Creating negative edge for test data

test_neg_edge_index = neg_edge_index[:, test_indices]
```

```python
test_edge_label_index = torch.cat(
            [testset.edge_index, test_neg_edge_index],
            dim=-1,
        )

test_edge_label = torch.cat([
            testset.edge_labels,
            neg_edge_labels[test_indices]
        ], dim=0)
```

```python

total_loss=[]
test_accuracies=[]
def test():
  model.eval()
  z=model.encode(torch.Tensor(test_adj_matrix),torch.Tensor(func_normalize(testset.x)),testset.edge_index)
  out = model.decode(z, test_edge_label_index).view(-1)
  #loss=criterion(out,test_edge_label).view(-1)
  loss = F.binary_cross_entropy_with_logits(out,test_edge_label)
  total_loss.append(loss.item())

  predictions = (torch.sigmoid(out) > 0.5).float()
  correct_predictions = (predictions == test_edge_label).float()
  acc = correct_predictions.mean()
  test_accuracies.append(acc)
  return acc
```

```python
for epoch in range(1,500):
  loss=train()
  acc=test()
  print(f"Epoch {epoch}, train_Loss: {loss:.10f}, test_Accuracy: {acc:.4f}")
```
